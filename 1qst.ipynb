```python
class node:
    def __init__(self, info=None, prox=None):
        self.info = info
        self.prox = prox

class listEncd:
    def __init__(self):
        self.ref = None

    def lobby(self):
        print("\n--- Menu Lista Encadeada ---")
        print("(i) Inserir\t(r) Remover\n(I) Imprimir\n(l)limpar memoria\n(f) Fechar")
        resp = input("\nDigite sua resposta: ")
        
        if resp == 'i':
            self.inserir()
        elif resp == 'I':
            self.imprimir()
        elif resp == 'l':
            self.limpar()
        elif resp == 'r':
            self.remover()
        elif resp != 'f':
            print("Opção não implementada ou inválida.")
            self.lobby()

    def inserir(self):
            valor = input("Digite o valor a ser inserido: ")
          # 1. Cria o novo nó com o valor digitado
            novo_no = node(valor)
            
            if not self.ref:
                novo_no.prox = self.ref
                self.ref = novo_no
            elif self.ref >= valor:
                if not self.ref:
                    novo_no.prox = self.ref
                    self.ref = novo_no    
            elif self.ref < valor :
                anterior = self.ref
                sucessor = self.ref.prox 
                while anterior > sucessor:
                    anterior.prox = sucessor.prox
                novo_no.prox = self.ref
                self.ref = novo_no
            else:
                atual = self.ref
                ponte = self.ref.prox
                while(ponte):
                    if ponte.info == valor:
                        atual.prox = ponte.prox
                        ponte.next = None
                atual = ponte
                ponte = ponte.next
            print(f"Valor {valor} inserido!")
        
            self.lobby()

    def limpar(self):
        #para Limpar a memória do objeto devemos simplesmente retirar a referência da listaencad
        self.ref = None
        self.lobby()
        print("Lista Limpada!!!!")

    def remover(self):
        #remove um elemento específico da lista
        valor = input("\nDigite o elemento que deseja remover: ")
        if not self.ref:
            print("Lista ja vazia!")
        elif self.ref == valor :
            self.ref = self.ref.prox
            print(f"\n{valor} foi removido !!!")
            self.lobby()
        else:
            atual = self.ref
            ponte = self.ref.prox
            while(ponte):
                if ponte.info == valor:
                    atual.prox = ponte.prox
                    ponte.next = None
                atual = ponte
                ponte = ponte.next
            print(f"\n{valor} foi removido !!!")
            self.lobby()

    def imprimir(self):
        atual = self.ref
        if not atual:
            print("Lista vazia!")
        else:
            while atual:
                print(f"[{atual.info}]", end=" -> ")
                atual = atual.prox
            print("None")
        self.lobby()

# Para rodar:
minha_lista = listEncd()
minha_lista.lobby()
```
